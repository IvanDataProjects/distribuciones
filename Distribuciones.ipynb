## Calcular Probabilidades

### Alturas

Graficar una distribución que represente las alturas de una población, la altura promedio es de 1.7 metros y el desvío estandar es de 0.1 metros.

```python
altura = np.random.normal(loc = 1.7, scale = 0.1, size = 1000000)
```

```python
sns.histplot(altura,stat="density")
```

Calcular probabilidad de que una persona mida más de 2 metros

```python
(altura > 2).sum() / 1000000
```

Calcular probabilidad de que una persona mida menos de 1.4 metros

```python
(altura < 1.4).sum() / len(altura)
```

Calcular probabilidad de que una persona mida entre 1.5 metros y 1.7 metros

```python
((altura > 1.5) & (altura < 1.7)).sum() / len(altura)
```

Calcular probabilidad de que una persona mida exactamente 1.7 metros

```python
(altura == 1.7).sum() / len(altura)
```

Calcular probabilidad de que una persona no mida entre 1.4 metros y 2 metros

```python
((altura < 1.4) | (altura > 2)).sum() / len(altura)
```

### Call center

Graficar una distribución que represente las llamadas realizadas por hora a un call center, el cual recibe 8 llamadas por hora

```python
llamadas = np.random.poisson(lam=8, size = 100000)
```

```python
sns.countplot(x = llamadas)
```

Calcular probabilidad de que llamen entre 9 y 12 personas

```python
((llamadas >= 9)|(llamadas <= 12)).sum()/len(llamadas)
```

Calcular probabilidad de que llamen menos de 2 personas

```python
(llamadas < 2).sum()/len(llamadas)
```

Calcular probabilidad de que llamen exactamente 8 personas

```python
(llamadas == 8).sum()/len(llamadas)
```

Calcular la probabilidad de que llamen más de 9 personas

```python
(llamadas > 9).sum()/100000
```

```python
llamadas.max()
```
